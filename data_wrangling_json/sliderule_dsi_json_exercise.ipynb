## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

## Problem 1

Find the 10 countries with most projects

```python
world_bank_df = pd.read_json('data/world_bank_projects.json')
```

```python
project_count = world_bank_df.groupby(['country_namecode']).size().reset_index(name='count')
top_countries = project_count.sort_values('count', ascending=False).head(10)
top_countries
```

## Problem 2

Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
#load and normalize json data
world_bank_str = json.load((open('data/world_bank_projects.json')))
project_themes = json_normalize(world_bank_str,'mjtheme_namecode',['id'])
project_themes.head(10)
```

```python
#It looks like some projects can have multiple themes, and there are duplicate rows. 
#To address this, we will only take unique rows.
project_themes_unique = project_themes.drop_duplicates()
#count project id by code
theme_count = pd.DataFrame(project_themes_unique.code.value_counts().reset_index())
theme_count.columns = ['code', 'count']
top_themes = theme_count.head(10)
top_themes
```

```python
#Finally, apply mapping of code to name to theme name
filter = (project_themes['name'] != '')
theme_key = project_themes[['code','name']][filter].drop_duplicates()
theme_key
pd.merge(top_themes, theme_key, on='code')
```

## Problem 3

```python
#load json and normalize to df
world_bank_str = json.load((open('data/world_bank_projects.json')))
project_themes = json_normalize(world_bank_str,'mjtheme_namecode',['id'])
```

```python
#create df with mapping of code to name and do a left join
filter = (project_themes['name'] != '')
theme_key = project_themes[['code','name']][filter].drop_duplicates()
theme_key
project_themes_complete = pd.merge(project_themes[['code','id']], theme_key, how='left', on='code')
project_themes_complete.head()
```

```python
#alternate solution using dictionary
world_bank_str = json.load((open('data/world_bank_projects.json')))
project_themes = json_normalize(world_bank_str,'mjtheme_namecode',['id'])
filter = (project_themes['name'] != '')
theme_key = project_themes[['code','name']][filter].drop_duplicates()
#create dictionary from dataframe and use for mapping
theme_dict = dict(zip(theme_key.code, theme_key.name))
project_themes.name = project_themes.code.map(theme_dict)
```
